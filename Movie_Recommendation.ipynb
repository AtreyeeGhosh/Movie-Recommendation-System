```python
import pandas as pd
import numpy as np
import pickle
```

```python
movies=pd.read_csv('/content/movierecommendations.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tags']=movies['genre']+movies['overview']
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec=cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vec)
```

```python
similar
```

```python
pickle.dump(new_df, open('/content/movies.pkl', 'wb'))
pickle.dump(similar, open('/content/similarity.pkl', 'wb'))
pickle.dump(cv,open('/content/vec.pkl','wb'))
```

```python
new_df = pickle.load(open('/content/movies.pkl', 'rb'))
similar = pickle.load(open('/content/similarity.pkl', 'rb'))
cv = pickle.load(open('/content/vec.pkl', 'rb'))
```

```python
def recommend(movies):
  if movies not in new_df['title'].values:
        return ["Movie not found."]
  index=new_df[new_df['title']==movies].index[0]
  distance=sorted(list(enumerate(similar[index])),reverse=True,key=lambda vec:vec[1])
  recommended_movies = []
  for i in distance[1:6]:
        recommended_movies.append(new_df.iloc[i[0]].title)
  return recommended_movies
```
